```python
'''Image Visualization
'''

import pprint
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
from skimage import io
import numpy as np

# helper func.
def show_image(path_list):
    row = len(path_list[0])
    col = len(path_list)
    fig = plt.figure(figsize=(row*5,col*8))
    fig.tight_layout()
    for j, data in enumerate(path_list):
        for i, path in enumerate(data):
            ax = plt.subplot2grid((col,row), (j,i))
            ax.spines['bottom'].set_linewidth(5.5)
            ax.spines['top'].set_linewidth(5.5)
            ax.spines['left'].set_linewidth(5.5)
            ax.spines['right'].set_linewidth(5.5)
            if i == 0:
                ax.spines['bottom'].set_color('green')
                ax.spines['top'].set_color('green')
                ax.spines['left'].set_color('green')
                ax.spines['right'].set_color('green')
            else:
                ax.spines['bottom'].set_color('red')
                ax.spines['top'].set_color('red')
                ax.spines['left'].set_color('red')
                ax.spines['right'].set_color('red')
                    
            ax.set_xticks([])
            ax.set_yticks([])
            im = get_image_from_path(path)
            plt.imshow(im)
    plt.show()


def get_image_from_path(image_path, crop_bbox=None):
    # download the image, convert it to a NumPy array, and then read
    # it into OpenCV format
    try:
        img = io.imread(image_path)
    except Exception as e:
        print("exception:", e)
        img = None

    if img is None or len(img.shape) > 3:
        img = np.zeros((1, 1, 3))
    try:
        img = img[:, :, :3]
    except Exception as e:
        print(img.shape)
        img = np.zeros((1, 1, 3))

    if crop_bbox:
        x_st = crop_bbox[0]
        y_st = crop_bbox[1]
        x_en = crop_bbox[2]
        y_en = crop_bbox[3]
        crop_img = img[y_st:y_en, x_st:x_en]
        img = crop_img

    return img

# Main
image_path1 = [
['/home1/irteam/users/nashory/dataset/nfd-cc-indiv/395639f6dffb4fed826127a3faf8af6b/395639f6dffb4fed826127a3faf8af6b-b2234806.jpg','/home1/irteam/users/nashory/dataset/nfd-cc-indiv/f4086669c1f04ee883a75e85e0dbc9e2/f4086669c1f04ee883a75e85e0dbc9e2-b83b0f4c.jpg', '/home1/irteam/users/nashory/dataset/nfd-cc-indiv/f4086669c1f04ee883a75e85e0dbc9e2/f4086669c1f04ee883a75e85e0dbc9e2-d6b07270.jpg', '/home1/irteam/users/nashory/dataset/nfd-cc-indiv/2edef5a8d2b24a3c8dc39138024a4293/2edef5a8d2b24a3c8dc39138024a4293-0dd5173d.jpg', '/home1/irteam/users/nashory/dataset/nfd-cc-indiv/f4086669c1f04ee883a75e85e0dbc9e2/f4086669c1f04ee883a75e85e0dbc9e2-9effea97.jpg', '/home1/irteam/users/nashory/dataset/nfd-cc-indiv/2edef5a8d2b24a3c8dc39138024a4293/2edef5a8d2b24a3c8dc39138024a4293-1103eeae.jpg']
]
image_path2 = [
['/home1/irteam/users/nashory/dataset/nfd-cc-indiv/bac5c5fc2d6549f583816d35344897e8/bac5c5fc2d6549f583816d35344897e8-6363cbb2.jpg','/home1/irteam/users/nashory/dataset/nfd-cc-indiv/14473a8d9c5b4492bac4af2a214ac2fc/14473a8d9c5b4492bac4af2a214ac2fc-2c0cccf3.jpg', '/home1/irteam/users/nashory/dataset/nfd-cc-indiv/f7a9e232f2354b5d8acec3e84fe3c50a/f7a9e232f2354b5d8acec3e84fe3c50a-b111d3da.jpg', '/home1/irteam/users/nashory/dataset/nfd-cc-indiv/14473a8d9c5b4492bac4af2a214ac2fc/14473a8d9c5b4492bac4af2a214ac2fc-9adb5cf3.jpg', '/home1/irteam/users/nashory/dataset/nfd-cc-indiv/395639f6dffb4fed826127a3faf8af6b/395639f6dffb4fed826127a3faf8af6b-619bde1c.jpg', '/home1/irteam/users/nashory/dataset/nfd-cc-indiv/f7a9e232f2354b5d8acec3e84fe3c50a/f7a9e232f2354b5d8acec3e84fe3c50a-4531ea27.jpg']
]
show_image(image_path2)
```
