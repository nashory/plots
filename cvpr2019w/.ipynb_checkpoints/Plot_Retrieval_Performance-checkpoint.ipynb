```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np

SMALL_SIZE = 20
MEDIUM_SIZE = 24
BIGGER_SIZE = 28

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title

fig = plt.figure()
plt.figure(figsize=(16,12))
ax = plt.axes()
ax.set(xlabel='# Retrieved Images (k=1, ..., 50)',
       ylabel='Retrieval Accuracy (Recall@k)');

x = [ 1, 5, 10, 20, 30, 40, 50]
y = dict()
y['ResNet50-SPoC+DF'] = [ 0.884, 0.960, 0.975, 0.983, 0.987, 0.990, 0.991 ]
y['ResNet50-SPoC+NFD'] = [ 0.716, 0.826, 0.861, 0.894, 0.909, 0.919, 0.925 ]
y['ResNet50-SPoC+ImgNet'] = [ 0.411, 0.572, 0.636, 0.693, 0.727, 0.749, 0.766 ]
y['FashionNet'] = [ 0.533, 0.673, 0.726, 0.764, 0.782, 0.791, 0.793 ]
y['FashionNet+Poselets'] = [ 0.424, 0.600, 0.653, 0.699, 0.720, 0.724, 0.736 ]
y['FashionNet+Joints'] = [ 0.414, 0.605, 0.641, 0.682, 0.715, 0.724, 0.723 ]
y['DARN'] = [ 0.381, 0.547, 0.620, 0.675, 0.700, 0.711, 0.716 ]


plt.plot(np.asarray(x), np.asarray(y['ResNet50-SPoC+DF']), '-ro', label='ResNet50-SPoC+DF(c)', lw=4, ms=10)
plt.plot(np.asarray(x), np.asarray(y['ResNet50-SPoC+NFD']), '-.gx', label='ResNet50-SPoC+NFD(c)', lw=4, ms=10)
plt.plot(np.asarray(x), np.asarray(y['ResNet50-SPoC+ImgNet']), '--bs', label='ResNet50-SPoC+ImgNet(c)', lw=4, ms=10)
plt.plot(np.asarray(x), np.asarray(y['FashionNet']), '--d', label='FashionNet*', lw=4, ms=10)
plt.plot(np.asarray(x), np.asarray(y['FashionNet+Poselets']), ':cv', label='FashionNet+Poselets*', lw=4, ms=10)
plt.plot(np.asarray(x), np.asarray(y['FashionNet+Joints']), '-.mP', label='FashionNet+Joints*', lw=4, ms=10)
plt.plot(np.asarray(x), np.asarray(y['DARN']), '-y^', label='DARN*', lw=4, ms=10)


plt.xlim(0, 50)
plt.ylim(0.4, 1.0);

plt.legend();
plt.grid(True);

# ax.plot(x, np.sin(x));

# plt.plot(x, x + 0, linestyle='solid')
# plt.plot(x, x + 1, linestyle='dashed')
# plt.plot(x, x + 2, linestyle='dashdot')
# plt.plot(x, x + 3, linestyle='dotted');

# # For short, you can use the following codes:
# plt.plot(x, x + 4, linestyle='-')  # solid
# plt.plot(x, x + 5, linestyle='--') # dashed
# plt.plot(x, x + 6, linestyle='-.') # dashdot
# plt.plot(x, x + 7, linestyle=':');  # dotted
```
