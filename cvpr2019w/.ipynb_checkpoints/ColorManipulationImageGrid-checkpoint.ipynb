```python
'''Image Visualization
'''

import pprint
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
from skimage import io
import numpy as np

# helper func.
def show_image(path_list):
    row = len(path_list[0])
    col = len(path_list)
    plt.figure(figsize=(row*5,col*5))
    for j, data in enumerate(path_list):
        for i, path in enumerate(data):
            ax = plt.subplot2grid((col,row), (j,i))
            ax.spines['bottom'].set_linewidth(5)
            ax.spines['top'].set_linewidth(5)
            ax.spines['left'].set_linewidth(5)
            ax.spines['right'].set_linewidth(5)
            if i == 0:
                ax.spines['bottom'].set_color('green')
                ax.spines['top'].set_color('green')
                ax.spines['left'].set_color('green')
                ax.spines['right'].set_color('green')
            else:
                ax.spines['bottom'].set_color('red')
                ax.spines['top'].set_color('red')
                ax.spines['left'].set_color('red')
                ax.spines['right'].set_color('red')
                    
            ax.set_xticks([])
            ax.set_yticks([])
            im = get_image_from_path(path)
            plt.imshow(im)
    plt.show()


def get_image_from_path(image_path, crop_bbox=None):
    # download the image, convert it to a NumPy array, and then read
    # it into OpenCV format
    try:
        img = io.imread(image_path)
    except Exception as e:
        print("exception:", e)
        img = None

    if img is None or len(img.shape) > 3:
        img = np.zeros((1, 1, 3))
    try:
        img = img[:, :, :3]
    except Exception as e:
        print(img.shape)
        img = np.zeros((1, 1, 3))

    if crop_bbox:
        x_st = crop_bbox[0]
        y_st = crop_bbox[1]
        x_en = crop_bbox[2]
        y_en = crop_bbox[3]
        crop_img = img[y_st:y_en, x_st:x_en]
        img = crop_img

    return img

# Main
image_path = [
['/home1/irteam/users/nashory/dataset/nfd-cc-indiv/539f37ef2e154425804964228472d311/539f37ef2e154425804964228472d311-6b345caf.jpg','/home1/irteam/users/nashory/dataset/nfd-cc-indiv/87f01524062a48f0829986d35da3eed9/87f01524062a48f0829986d35da3eed9-07625f12.jpg', '/home1/irteam/users/nashory/dataset/nfd-cc-indiv/bc576cb4035149ddb1c54bd9a7dcae71/bc576cb4035149ddb1c54bd9a7dcae71-af0cd090.jpg', '/home1/irteam/users/nashory/dataset/nfd-cc-indiv/1072cccd1f3143bba3d19e5f15911072/1072cccd1f3143bba3d19e5f15911072-6eb874c6.jpg', '/home1/irteam/users/nashory/dataset/nfd-cc-indiv/2dce96425e704217a84c380f98d79e9a/2dce96425e704217a84c380f98d79e9a-f10eb36d.jpg', '/home1/irteam/users/nashory/dataset/nfd-cc-indiv/4db2efe0463d4e1399cdd206608094f9/4db2efe0463d4e1399cdd206608094f9-1aa4b4ab.jpg', '/home1/irteam/users/nashory/dataset/nfd-cc-indiv/11c23f87e4ef47ce92be05722c04cfeb/11c23f87e4ef47ce92be05722c04cfeb-aa67ed93.jpg', '/home1/irteam/users/nashory/dataset/nfd-cc-indiv/4db2efe0463d4e1399cdd206608094f9/4db2efe0463d4e1399cdd206608094f9-fe7e1467.jpg', '/home1/irteam/users/nashory/dataset/nfd-cc-indiv/bc576cb4035149ddb1c54bd9a7dcae71/bc576cb4035149ddb1c54bd9a7dcae71-5cef3282.jpg', '/home1/irteam/users/nashory/dataset/nfd-cc-indiv/112c6b1a77284cc3a2cb16f0710ba821/112c6b1a77284cc3a2cb16f0710ba821-45aad297.jpg', '/home1/irteam/users/nashory/dataset/nfd-cc-indiv/2dfa0f043e3a4bc8a044f32ec0122e13/2dfa0f043e3a4bc8a044f32ec0122e13-b0706e65.jpg'],
]
show_image(image_path)



```
